```python
import speech_recognition as sr
import gradio as gr

# Initialize recognizer
recognizer = sr.Recognizer()

def recognize_speech_from_mic(audio_file):
    try:
        with sr.AudioFile(audio_file) as source:
            recognizer.adjust_for_ambient_noise(source)
            audio = recognizer.listen(source)
        
        text = recognizer.recognize_google(audio, language='ar-en')
        return f"You said: {text}"
    
    except sr.UnknownValueError:
        return "Sorry, I could not understand the audio."
    
    except sr.RequestError:
        return "Sorry, I could not request results from Google."

# Gradio interface
iface = gr.Interface(
    fn=recognize_speech_from_mic,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="Arabic-English Speech Recognition",
    description="Speak in Arabic or English, and this app will transcribe it using Google's speech recognition API."
)

iface.launch()
```
